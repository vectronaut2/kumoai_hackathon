```python
import pandas as pd
```

```python
data_path = "../../datasets/fb15k-237/"
```

```python
# triplets_train = pd.read_csv(f"{data_path}/derived/triplets_with_src_dst_wiki_ids.csv")
test = False
if test:
    triplets = pd.read_csv(f"{data_path}/derived/triplets_with_src_dst_wiki_ids.test.csv")
else:
    triplets = pd.read_csv(f"{data_path}/derived/triplets_with_src_dst_wiki_ids.csv")
```

```python
triplets.head()
```

```python
freebase_vs_wiki_df =  pd.read_csv(f"{data_path}/derived/fb_wiki_mapping_incl_wiki_extracts.csv")
```

```python
freebase_vs_wiki_df['wikidata_int_id'] = freebase_vs_wiki_df.apply(lambda row: int(row['wikidata_id'][1:]), axis=1)

freebase_vs_wiki_df.drop('wikidata_id', axis=1, inplace=True)
freebase_vs_wiki_df.rename(columns={'wikidata_int_id':'wikidata_id'}, inplace=True)
```

```python
freebase_vs_wiki_df.head()
```

```python
with_src_wiki_extract_df = pd.merge(
    triplets, freebase_vs_wiki_df[['wikidata_id', 'Title', 'Extract']], 
    left_on='src_wikidata_id', right_on='wikidata_id'
)
```

```python
with_src_dest_wiki_extract_df = pd.merge(
    with_src_wiki_extract_df, freebase_vs_wiki_df[['wikidata_id', 'Title', 'Extract']], 
    left_on='dest_wikidata_id', right_on='wikidata_id'
)
```

```python
with_src_dest_wiki_extract_df.info()
```

```python
with_src_dest_wiki_extract_df.head(3)
```

```python
with_src_dest_wiki_extract_df.drop(columns=['freebase_id_x', 'freebase_id_y', 'wikidata_id_x', 'wikidata_id_y'],inplace=True)
```

```python
with_src_dest_wiki_extract_df.rename(columns={
    'label_x':'src_wikidata_label','label_y':'dest_wikidata_label',
    'Title_x': 'src_wikidata_title','Title_y': 'dest_wikidata_title',
    'Extract_x':'src_wikidata_extract','Extract_y':'dest_wikidata_extract',
},inplace=True)
```

```python
if test:
    with_src_dest_wiki_extract_df.to_csv(f"{data_path}/derived/triplets_with_src_dst_wiki_extracts.test.csv")
else:
    with_src_dest_wiki_extract_df.to_csv(f"{data_path}/derived/triplets_with_src_dst_wiki_extracts.csv")
```

```python
# load generated file
if test:
    with_src_dest_wiki_extract_df = pd.read_csv(f"{data_path}/derived/triplets_with_src_dst_wiki_extracts.test.csv")
else:
    with_src_dest_wiki_extract_df = pd.read_csv(f"{data_path}/derived/triplets_with_src_dst_wiki_extracts.csv")
```

```python
prefix = f"{data_path}/derived/triplets_with_src_dst_wiki_extracts"
```

```python
# split file into 10k chunks each
k = 10000
df = with_src_dest_wiki_extract_df

num_rows = len(df)
for i in range(0, num_rows, k):
    chunk = df.iloc[i:i + k]
    print(len(chunk))
    filename = f"{prefix}.{i}.csv"
    chunk.to_csv(filename, index=False)
    
```
